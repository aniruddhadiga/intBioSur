```python
import numpy as np
np.random.seed(0) # Set seed for reproducibility
import pandas as pd
import matplotlib.pyplot as plt
import os
from IPython.display import Image

import ewstools
from ewstools.models import simulate_ricker
```

```python
series = simulate_ricker(tmax=1000, F=[0,2.7])
```

```python
series.plot()
```

```python
ts = ewstools.TimeSeries(data=series)
```

```python
ts.detrend(method='Lowess',span=14)
```

```python
ts.state['smoothing'].plot()
```

```python
ts.state['smoothing'].plot()
```

```python
cdf=pd.read_csv('/project/biocomplexity/COVID-19_commons/cdc_forecast_ensemble/input/case_data_cleaned/daily_case_data.csv',dtype={'FIPS':str})
cdf=cdf.set_index('FIPS')
```

```python
data=cdf.sum()
```

```python
ts = ewstools.TimeSeries(data=data)
```

```python
data.index=pd.to_datetime(data.index)
```

```python
rw=12
ts.compute_var(rolling_window=rw)
ts.compute_cv(rolling_window=rw)
ts.compute_auto(rolling_window=rw, lag=1)
ts.compute_skew(rolling_window=rw)
```

```python
_,ax=plt.subplots()
ts.ews['ac1'].plot(ax=ax)

ax2=ax.twinx()
ts.state.plot(ax=ax2,color='k')
ax.grid('on',which='both')
```

```python
_,ax=plt.subplots()
ts.ews['skew'].plot(ax=ax)

ax2=ax.twinx()
ts.state.plot(ax=ax2,color='k')
ax.grid('on',which='both')
```
