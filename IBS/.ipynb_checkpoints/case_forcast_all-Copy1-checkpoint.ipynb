```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import os
import numpy as np
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
import tensorflow as tf
import math
from statsmodels.tsa.api import SARIMAX, AutoReg
from statsmodels.tsa.arima.model import ARIMA

ww = pd.read_csv("/Users/cc/Downloads/Sewershed_PCR_Conc.csv",parse_dates=['sample_collect_date'])
cases = pd.read_csv("/Users/cc/Downloads/Sewershed_Case_Counts.csv",parse_dates=['event_date'])
ww['log_viral'] = ww['viral_load'].apply(lambda x: math.log(x+1, 10))
ww.describe()
```

```python
#deep learning settings
MAX_EPOCHS = 200

def compile_and_fit(model, train_x, train_y, patience=2):
  early_stopping = tf.keras.callbacks.EarlyStopping(monitor='mean_absolute_error',
                                                    patience=patience,
                                                    mode='min')

  model.compile(loss=tf.keras.losses.MeanSquaredError(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=[tf.keras.metrics.MeanAbsoluteError()])

  history = model.fit(train_x, train_y, epochs=MAX_EPOCHS, verbose = 0,
                      callbacks=[early_stopping])
  return history
```

```python
def linear_regression(trainx, trainy, testx):
    reg = LinearRegression().fit(np.asarray(trainx), np.asarray(trainy))
    predicted_cases = reg.predict(np.array(testx))
    predicted_cases[predicted_cases < 0] = 0
    return predicted_cases

def arimax_model(trainx, trainy, testx, test_start, test_end):
    arima_exog_res = ARIMA(trainy, exog=trainx, order=(1, 0, 0), trend="c").fit()
    return arima_exog_res.predict(start = test_start, end = test_end, exog = testx)

def arima_model(trainy, test_start, test_end):
    arima_res = ARIMA(trainy, order=(1, 0, 0), trend="c").fit()
    return arima_res.predict(start = test_start, end = test_end)
    
def twolayerMLP(trainx, trainy, testx, reps):
    results = []
    for i in range(reps):
        two_layer = tf.keras.Sequential([
            tf.keras.layers.Dense(units=3, activation='relu'),
            tf.keras.layers.Dense(units=4, activation='relu'),
            tf.keras.layers.Dense(units=1)
        ])
        history = compile_and_fit(two_layer, trainx, trainy)
        predicts = two_layer.predict(testx)
        results.append(predicts)
    # get mean results
    ntest = len(testx)
    sum_predicts = [0]*(ntest)
    for i in range(reps):
        for j in range(ntest):
            sum_predicts[j] += results[i][j][0]
        #print(sum_predicts)
    mean_results = [y/reps for y in sum_predicts ]
    return mean_results

def construct_multistep_sample(features, labels, step):
    stacked_features = []
    total_len = len(labels)
    for i in range(step,total_len):
        group_feature = features[i-step: i]
        stacked_features.append(np.array([np.array(x) for x in group_feature]))
    return (np.array(stacked_features), np.array(labels[step:total_len]))  

def multidense(trainx, trainy, testx, reps):
    results = []
    for i in range(reps):
        multidense_model = tf.keras.models.Sequential([
            # Shape [batch, time, features] => [batch, time, lstm_units]
            ##tf.keras.layers.LSTM(32, return_sequences=True),
            # Shape => [batch, time, features]
            ##tf.keras.layers.Dense(units=1)
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(units=32, activation='relu'),
            tf.keras.layers.Dense(units=32, activation='relu'),
            tf.keras.layers.Dense(units=1),
            # Add back the time dimension.
            # Shape: (outputs) => (1, outputs)
            #tf.keras.layers.Reshape([1, -1]),
        ])

        history = compile_and_fit(multidense_model, trainx, trainy)
        predicts = multidense_model.predict(testx)
        results.append(predicts)
    ntest = len(testx)
    sum_predicts = [0]*ntest
    for i in range(reps):
        for j in range(ntest):
            sum_predicts[j] += results[i][j][0]
    mean_results = [y/reps for y in sum_predicts ]
    return mean_results


def lstm_model(trainx, trainy, testx, step, reps):
    results = []
    for i in range(reps):
        lstm = tf.keras.models.Sequential([
            # Shape [batch, time, features] => [batch, time, lstm_units]
            tf.keras.layers.LSTM(32, return_sequences=True),
            tf.keras.layers.LSTM(32, return_sequences=True),
            # Shape => [batch, time, features]
            tf.keras.layers.Dense(units=1)
            # Add back the time dimension.
            # Shape: (outputs) => (1, outputs)
            #tf.keras.layers.Reshape([1, -1]),
        ])

        history = compile_and_fit(lstm, trainx, trainy)
        predicts = lstm.predict(testx)
        results.append(predicts)
    ntest = len(testx)
    sum_predicts = [0]*ntest
    for i in range(reps):
        for j in range(ntest):
            sum_predicts[j] += results[i][j][step-1][0]
    mean_results = [y/reps for y in sum_predicts ]
    return mean_results  
```

```python
def construct_ww_features(features, labels, step):
    stacked_features = []
    total_len = len(labels)
    for i in range(step,total_len):
        ww_feature = []
        for j in range(step):
            ww_feature.append(features[i-step+j][0])
        stacked_features.append(np.array(ww_feature))
    return (np.array(stacked_features), np.array(labels[step:total_len])) 


def construct_cases_features(features, labels, step):
    stacked_features = []
    total_len = len(labels)
    for i in range(step,total_len):
        cases_feature = []
        for j in range(step):
            cases_feature.append(features[i-step+j][1])
        stacked_features.append(np.array(cases_feature))
    return (np.array(stacked_features), np.array(labels[step:total_len])) 
```

```python
# assign sewershed name here
sewershed = 'Virginia Initiative Plant'
ww['sample_collect_date'] = ww['sample_collect_date'].apply(lambda x: pd.Timestamp(x))
cases['event_date'] = cases['event_date'].apply(lambda x: pd.Timestamp(x))
ww_data = ww[ww['sewershed'] == sewershed]
case_data = cases[cases['sewershed'] == sewershed]
ww_viral = ww_data.set_index('sample_collect_date')['viral_load']
ww_viral_log = ww_data.set_index('sample_collect_date')['log_viral']
case_plot = case_data.set_index('event_date')['count']
case_plot.sort_index(inplace=True)
ww_viral.sort_index(inplace=True)
ww_viral_log.sort_index(inplace=True)

# aggregate cases by week
viral_dict = ww_viral_log.to_dict()
case_dict = {}
for rec in viral_dict:
    start_time = rec.to_pydatetime() - dt.timedelta(days=6)
    dated_cases = case_plot.loc[pd.Timestamp(start_time): rec]
    case_dict[rec] = dated_cases.sum()
#print((case_dict))
```

```python
features = []
targets = []
viral_list = list(viral_dict.keys())
for i in range(len(viral_list)-1):
    viral_date = viral_list[i]
    next_viral_date = viral_list[i+1]
    date_diff = (next_viral_date.to_pydatetime()-viral_date.to_pydatetime()).days
    if date_diff > 10:
        continue
    case_date = next_viral_date
    feature = [viral_dict[viral_date], case_dict[viral_date]]
    if ww_data[ww_data['sample_collect_date'] == viral_date]['pcr_target_below_lod'].bool():
        feature.append(1)
    else:
        feature.append(0)
    targets.append(case_dict[case_date])
    features.append(feature)

print(len(targets))
```

```python
### construct training the test set, the index need to be adjusted by need
#2021.9-2022.3
train_x = features[:21]
train_y = targets[:21]

#2022.10-2022.12
train2_x = features[-17:-6]
train2_y = targets[-17:-6]

#2023.1
test_x = features[-6:-1]
test_y = targets[-6:-1]
test_x3 = features[-9:-1]
test_y3 = targets[-9:-1]

# construct training set for arima model
exo_train = [x[0] for x in train_x]
exo_test = [x[0] for x in features[21:66]]
exo_train2 = [x[0] for x in train2_x]
exo_test2 = [x[0] for x in features[-6:]]

# convert the training and test set into tensorflow input format
nptrain_x=np.array([np.array(xi) for xi in train_x])
nptrain_y = np.array(train_y)

nptrain2_x=np.array([np.array(xi) for xi in train2_x])
nptrain2_y = np.array(train2_y)

nptest_x = np.array([np.array(xi) for xi in test_x])
nptest_y = np.array(test_y)

nptrain_x3, nptrain_y3 = construct_multistep_sample(train_x, train_y, 3)
nptest_x3, nptest_y3 = construct_multistep_sample(test_x3, test_y3, 3)

# construct wastewater only feature from past three weeks, the parameter can be adjusted
ww_train, ww_y = construct_ww_features(train_x, train_y, 3)
ww_train2, ww_y2 = construct_ww_features(train2_x, train2_y, 3)
ww_test, ww_testy = construct_ww_features(test_x3, test_y3, 3)

# construct case only feature from past three weeks, the parameter can be adjusted
cases_train, cases_y = construct_cases_features(train_x, train_y, 3)
cases_train2,cases_y2 = construct_cases_features(train2_x, train2_y, 3)
cases_test, cases_testy = construct_cases_features(test_x3, test_y3, 3)
print(cases_test, cases_testy, test_y)
```

```python
def evaluate(predict, testy):
    mae = metrics.mean_absolute_error(testy, predict)
    mape = metrics.mean_absolute_percentage_error(testy, predict)
    mse = metrics.mean_squared_error(testy, predict)
    rmse = np.sqrt(mse) # or mse**(0.5) 
    return(mae, mape, rmse)
```

```python
predict_lin_old = linear_regression(train_x, train_y, test_x)
predict_lin_12w = linear_regression(train2_x, train2_y, test_x)
predict_2layer_old = twolayerMLP(nptrain_x, nptrain_y, nptest_x, 10)
predict_mdense_old = multidense(nptrain_x3, nptrain_y3, nptest_x3, 10)
predict_lstm_old = lstm_model(nptrain_x3, nptrain_y3, nptest_x3, 3, 10)
predict_arimax_old = arimax_model(exo_train, train_y, exo_test, 60, 65)
predict_arimax_12w = arimax_model(exo_train2, train2_y, exo_test2, 12, 16)
predict_linww_old = linear_regression(ww_train, ww_y, ww_test)
predict_linww_12w = linear_regression(ww_train2, ww_y2, ww_test)
```

```python
#case only baselines
predict_arima_old = arima_model(train_y, 60, 65)
predict_arima_12w = arima_model(train2_y, 12, 16)
predict_lincase_old = linear_regression(cases_train, cases_y, cases_test)
predict_lincase_12w = linear_regression(cases_train2, cases_y2, cases_test)
```

```python
print(evaluate(predict_lin_old, test_y))
print(evaluate(predict_lin_12w, test_y))
print(evaluate(predict_2layer_old, test_y))
print(evaluate(predict_mdense_old, test_y))
print(evaluate(predict_lstm_old, test_y))
print(evaluate(predict_arimax_old[:5], test_y))
print(evaluate(predict_arimax_12w[:5], test_y))
print(evaluate(predict_linww_old, test_y))
print(evaluate(predict_linww_12w, test_y))
print(evaluate(predict_arima_old[:5], test_y))
print(evaluate(predict_arima_12w[:5], test_y))
print(evaluate(predict_lincase_old, test_y))
print(evaluate(predict_lincase_12w, test_y))
```
