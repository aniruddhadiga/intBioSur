```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import os
import numpy as np
import seaborn as sns
%config Completer.use_jedi = False
```

```python
ww = pd.read_csv("/project/biocomplexity/COVID-19_commons/data/opioids_url/Sewershed_PCR_Conc.csv")#,parse_dates=['sample_collect_date'])
cases = pd.read_csv("/project/biocomplexity/COVID-19_commons/data/opioids_url/Sewershed_Case_Counts.csv",parse_dates=['event_date'])
```

```python
ssh_df=pd.read_csv('/project/biocomplexity/COVID-19_commons/data/VDH_wastewater/county_mappings/ssh_name_mapping.csv')
ssh_map=dict(zip(ssh_df['new_file_name'],ssh_df['dashboard_name']))
```

```python
def pcr_to_viral(pcr_conc, flow_rate):
    return pcr_conc*flow_rate*3785411.8

def recover_viral_load(raw_load, rec_percent):
    if rec_percent == 0:
        return raw_load
    return raw_load/rec_percent
```

```python
sewershed = 'ARenew'
date_replace_dict = {'20222-05-2':'2022-05-02', '20021-09-2':'2021-09-02'}
ww.replace({'sample_collect_date': date_replace_dict },inplace=True)
ww['sample_collect_date'] = ww['sample_collect_date'].apply(lambda x: pd.Timestamp(x))
cases['event_date'] = cases['event_date'].apply(lambda x: pd.Timestamp(x))
ww_data = ww[ww['sewershed'] == sewershed]
case_data = cases[cases['sewershed'] == sewershed]
#ww_data['VIRAL LOAD'] = pcr_to_viral(ww_data['pcr_target_avg_conc'], ww_data['pcr_target_avg_conc'])
ww_viral = ww_data.set_index('sample_collect_date')['viral_load']
ww_pcr = ww_data.set_index('sample_collect_date')['pcr_target_avg_conc']
case_plot = case_data.set_index('event_date')['count']
case_plot.sort_index(inplace=True)
ww_viral.sort_index(inplace=True)
ww_pcr.sort_index(inplace=True)
```

```python
# plot sero curves
wwtp = sewershed
plt.figure(figsize=(20,4))
ax1 = plt.gca()
ax2 = ax1.twinx()
case_label = wwtp + ' case'
ax1.plot(case_plot,'r-*')
ww_label = wwtp + ' Viral Load'
ax2.plot(ww_viral,'b-^')
ax1.legend([case_label], loc='upper right',fontsize=12)
ax1.set_ylabel('WWTP Cases')
ax2.set_ylabel('Wastewater')
ax2.legend([ww_label],loc='upper left',fontsize=12)
ax1.set_xlabel('Week')
ax2.set_xlabel('Week')
title = wwtp + ' Wastewater vs. cases'
plt.title(title,fontsize=14)
plt.show()
```
