```python
def gen_data_cc(sewershed,ww,cases):
    date_replace_dict = {'20222-05-2':'2022-05-02', '20021-09-2':'2021-09-02'}
    # ww.replace({'sample_collect_date': date_replace_dict },inplace=True)
    ww['sample_collect_date'] = ww['sample_collect_date'].apply(lambda x: pd.Timestamp(x))
    cases['event_date'] = cases['event_date'].apply(lambda x: pd.Timestamp(x))
    ww_data = ww[ww['sewershed'] == sewershed]
    case_data = cases[cases['sewershed'] == sewershed]
    #ww_data['VIRAL LOAD'] = pcr_to_viral(ww_data['pcr_target_avg_conc'], ww_data['pcr_target_avg_conc'])
    ww_viral = ww_data.set_index('sample_collect_date')['viral_load']
    ww_pcr = ww_data.set_index('sample_collect_date')['pcr_target_avg_conc']
    case_plot = case_data.set_index('event_date')['count']
    case_plot.sort_index(inplace=True)
    ww_viral.sort_index(inplace=True)
    ww_pcr.sort_index(inplace=True)

    viral_dict = ww_viral.to_dict()
    case_dict = {}
    for rec in viral_dict:
        start_time = rec.to_pydatetime() - dt.timedelta(days=6)
        dated_cases = case_plot.loc[pd.Timestamp(start_time): rec]
        case_dict[rec] = dated_cases.sum()

    temp1=pd.Series(case_dict, name='cases')
    temp1.index.name='Attribute'
    temp1=temp1.reset_index()
    temp1.loc[:,'sewershed']=sewershed

    temp2=pd.Series(viral_dict, name='ww')
    temp2.index.name='Attribute'
    temp2=temp2.reset_index()
    temp2.loc[:,'sewershed']=sewershed
    temp=temp1.merge(temp2)
    return temp
```

```python
all_data=pd.DataFrame()
for s in tr_swsh:
    try:
        all_data=all_data.append(gen_data_cc(s,ww,cases),sort=True)
    except:
        print(s)
        continue
```

```python
dt=pd.to_datetime('2022-01-01')
corrdf=pd.DataFrame(columns=['corr','lag'])
for loc in ['Richmond','Alexandria Renew']:
    dates=[]
    for j in range(0,55):
        temp=pd.DataFrame(columns=['ww','cases'],index=pcases.index)
        temp.loc[:,'ww']=pww.loc[:,loc]
        ind=temp.index
        temp_corr=0
        temp_lag=0
        for i in range(-10,10):
            temp.loc[:,'cases']=pcases.loc[:,loc].shift(i)
            temp2=temp.loc[ind].dropna()
            stdt=dt+pd.Timedelta(weeks=j)
            enddt=stdt+pd.Timedelta(weeks=12)
            pcr_i=pcr(temp2.loc[stdt:enddt,'cases'],temp2.loc[stdt:enddt,'ww'])
            if (pcr_i[1]<=0.05) and pcr_i[0]>temp_corr:
                temp_lag=i
                temp_corr=pcr_i[0]
        corrdf.loc[stdt,loc+'corr']=temp_corr
        corrdf.loc[stdt,loc+'lag']=temp_lag
```
