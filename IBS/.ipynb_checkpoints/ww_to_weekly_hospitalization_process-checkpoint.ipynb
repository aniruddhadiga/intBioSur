```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import os
import numpy as np
from sklearn.linear_model import LinearRegression
import sklearn.metrics as metrics
import math
from statsmodels.tsa.api import SARIMAX, AutoReg
from statsmodels.tsa.arima.model import ARIMA
```

```python
fip_to_ww = pd.read_csv("/project/biocomplexity/COVID-19_commons/products/external_data_collection/VDH_wastewater/county_mappings/FIPS_to_WWS.csv", dtype={'FIPS':str})
valid_fip_to_ww = fip_to_ww[fip_to_ww['WWS']!='None'].reset_index()
fips_set = set(valid_fip_to_ww['FIPS'].unique())
```

```python
# Pull in raw hospitalization data by county
county_hosp_data_raw = pd.read_csv('/project/biocomplexity/COVID-19_commons/data/COVIDCast/dsew-cpr_confirmed_admissions_covid_1d_7dav_county.csv',parse_dates=['time_value','issue'],dtype={'geo_value':str})
county_hosp_data_raw = county_hosp_data_raw[['geo_value', 'time_value', 'issue', 'value']]
county_hosp_data_raw['geo_value'] = county_hosp_data_raw['geo_value'].str.zfill(5)

# filter for VA sewershed related data only
county_hosp_data_raw = county_hosp_data_raw[county_hosp_data_raw['geo_value'].isin(fips_set)]
#county_hosp_data_raw = county_hosp_data_raw[county_hosp_data_raw['geo_value'].astype(int) < 52000]

# adjust raw data to be weekly and line up with viral load Attribute dates
county_hosp_data_adj = county_hosp_data_raw.copy()
county_hosp_data_adj['value'] = county_hosp_data_adj['value'] * 7
county_hosp_data_adj['time_value_adj'] = county_hosp_data_adj['time_value'] - dt.timedelta(days=0)
county_hosp_data_adj = county_hosp_data_adj.groupby(['time_value_adj', 'geo_value']).sum().reset_index()

# rename columns
county_hosp_data_adj.columns = ['Attribute', 'FIPS', 'hosps']
county_hosp_data_adj
```

```python
hosps_to_ww = county_hosp_data_adj.merge(
    valid_fip_to_ww, left_on='FIPS', right_on='FIPS')
hosps_to_ww['adjusted_hosps'] = hosps_to_ww['hosps']*hosps_to_ww['Fraction']
hosps_all_site = hosps_to_ww.groupby(['Attribute', 'WWS']).agg({'adjusted_hosps': ['sum']})
hosps_all_site.columns = ['site_hosps']
hosps_all_site = hosps_all_site.reset_index()
hosps_all_site.sort_values(by = ['Attribute'], inplace = True)
hosps_all_site
```

```python
wwtp = 'HRSD: Onancock'
########################################
hosp_by_site = hosps_all_site[hosps_all_site['WWS'] == wwtp]
hosp_site_series = hosp_by_site.set_index('Attribute')['site_hosps']

# load wastewater and sewershed cases
ww = pd.read_csv("/Users/cc/Downloads/Sewershed_PCR_Conc.csv",parse_dates=['sample_collect_date'])
cases = pd.read_csv("/Users/cc/Downloads/Sewershed_Case_Counts.csv",parse_dates=['event_date'])
sewershed = 'Onancock'
########################################
date_replace_dict = {'20222-05-2':'2022-05-02', '20021-09-2':'2021-09-02'}
# ww.replace({'sample_collect_date': date_replace_dict },inplace=True)
ww['sample_collect_date'] = ww['sample_collect_date'].apply(lambda x: pd.Timestamp(x))
ww['log_viral'] = ww['viral_load'].apply(lambda x: math.log(x+1,10))
ww_data = ww[ww['sewershed'] == sewershed]
#ww_viral = ww_data.set_index('sample_collect_date')['log_viral']
ww_viral = ww_data.set_index('sample_collect_date')['viral_load']
ww_viral.sort_index(inplace=True)
cases['event_date'] = cases['event_date'].apply(lambda x: pd.Timestamp(x))
case_data = cases[cases['sewershed'] == sewershed]
case_plot = case_data.set_index('event_date')['count']
case_plot.sort_index(inplace=True)
```

```python
# merge hospital counts and wastewater
viral_dict = ww_viral.to_dict()
full_hosp_dict = hosp_site_series.to_dict()
#match viral date with nearest hosp data
hosp_dict = {}
for rec in viral_dict:
    if rec in full_hosp_dict:
        hosp_dict[rec] = full_hosp_dict[rec]
        continue
    for i in range(3):
        start_time = rec.to_pydatetime() + dt.timedelta(days=i+1)
        if pd.Timestamp(start_time) in full_hosp_dict:
            hosp_dict[rec] = full_hosp_dict[pd.Timestamp(start_time)]
            break
        start_time = rec.to_pydatetime() - dt.timedelta(days=i+1)
        if pd.Timestamp(start_time) in full_hosp_dict:
            hosp_dict[rec] = full_hosp_dict[pd.Timestamp(start_time)]
            break
print((hosp_dict))

# aggregate cases by week
case_dict = {}
for rec in viral_dict:
    start_time = rec.to_pydatetime() - dt.timedelta(days=6)
    dated_cases = case_plot.loc[pd.Timestamp(start_time): rec]
    case_dict[rec] = dated_cases.sum()
#print((case_dict))
```

```python
plt.figure(figsize=(20,4))
ax1 = plt.gca()
ax2 = ax1.twinx()
county_label = wwtp + ' county mapped hosps'
#ax1.plot(pd.Series(case_dict),'r-*')
#ax1.plot(test['site_hosps'], 'g-*')
ax1.plot(pd.Series(hosp_dict), 'g-*')
ww_label = wwtp + ' Viral Load'
#ax2.plot(test['viral_load'],'b-^')
ax2.plot(pd.Series(viral_dict),'b-^')
ax1.legend([county_label], loc='upper right',fontsize=12)
ax1.set_ylabel('WWTP Cases')
ax2.set_ylabel('Wastewater')
ax2.legend([ww_label],loc='upper left',fontsize=12)
ax1.set_xlabel('Week')
ax2.set_xlabel('Week')
title = wwtp + ' Wastewater vs. cases'
plt.title(title,fontsize=14)
plt.show()
```
